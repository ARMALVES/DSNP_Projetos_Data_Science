Naquela ocasião, um senhor de 61 anos tinha dado entrada no Hospital Israelita Albert Einstein em São Paulo no dia anterior, após uma viagem à Itália, na região da Lombardia.

De lá pra cá, dia em que foi finalizado este artigo (05 de julho de 2020), o vírus veio se apresentando cada vez mais aos brasileiros, se fazendo presente e mudando nossa rotina bruscamente de tal maneira que somos testados diariamente o quão forte nosso sistema é, seja físico ou mental.

Após 130 dias convivendo com essa nova realidade, conhecemos seus sintomas e seus desfechos que por muitas vezes são fatais e não permite um último adeus.
