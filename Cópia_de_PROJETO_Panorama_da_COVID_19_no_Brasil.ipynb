### 1.1- O que é a COVID-19

[COVID-19](https://pt.wikipedia.org/wiki/COVID-19) é uma doença infeciosa causada pelo coronavírus da Síndrome Respiratória Aguda Grave 2 (SARS-CoV-2), em inglês *Severe Acute Respiratory Syndrome*.

### 1.2- COVID-19 e sua origem

<p align="justify">A Organização Mundial de Saúde (OMS) considera que os morcegos são o reservatório natural mais provável do vírus, embora algumas diferenças entre os vírus encontrados em morcegos e os encontrados em seres humanos sugiram que os humanos foram infetados através de um hospedeiro intermédio. As primeiras infeções conhecidas foram descobertas na cidade de Wuhan (província de Hubei, China).</p>


### 1.3- Como a COVID-19 age no corpo humano

Os pulmões são o principal alvo da covid-19, mas a doença não afeta apenas o sistema respiratório. Coração, rins, nervos, cérebro e até a pele são atacados pelo patógeno, com danos potencialmente duradouros.

Abaixo deixo um resumo dos órgãos citados acima, atacados pela COVID-19. As informações completas estão no site [Deutsche Welle](https://www.dw.com/pt-br/como-o-novo-coronav%C3%ADrus-ataca-todo-o-organismo/a-53418039), emissora internacional da Alemanha.

* <u>**Coração**</u>

 Diversos estudos e artigos científicos dos Estados Unidos, China e Itália, entre outros países, sugerem que o Sars-cov-2 também ataque o coração. 

 Em pacientes anteriormente saudáveis, a infecção viral também tem causado inflamação do miocárdio. Não se sabe se quem causa esses danos é o próprio coronavírus ou – **o que parece mais provável** – as reações imunológicas desencadeadas pela infecção.

* <u>**Pulmões**</u>

  No decorrer da covid-19, os pulmões são fortemente atacados. Muitos pacientes apresentam como sequela uma redução da capacidade pulmonar.

  Pesquisa-se no momento se esses pacientes desenvolveram fibrose pulmonar. Ao enrijecer o tecido dos pulmões, ela pode dificultar o transporte de oxigênio para os vasos sanguíneos, provocando em respiração superficial e rápida, e tosse seca e irritável. O resultado é uma queda do desempenho físico, tornando difíceis até mesmo atividades quotidianas.

  A fibrose pulmonar é irreversível, já que as cicatrizes no órgão não retrocedem, porém seu progresso pode ser desacelerado ou mesmo sustado, se ela é detectada a tempo.

* <u>**Sistema vascular**</u>

  Na autópsia de vítimas da covid-19, patologistas do Hospital Universitário de Zurique descobriram que em alguns deles estava inflamada toda a camada celular interna dos vasos sanguíneos e linfáticos (endotélio) de alguns órgãos.

  A conclusão foi que o vírus Sars-cov-2 causa uma inflamação generalizada do endotélio. A consequência podem ser em falência de múltiplos órgãos e morte.

* <u>**Sistema nervoso**</u>

  Em mais de 80% dos pacientes de covid-19, observa-se, logo no início da infecção, uma redução dos sentidos de paladar e olfato. Esse é um dos sintomas considerados no diagnóstico precoce, diferente de um resfriado comum ou conjuntivite que só surge num estágio avançado da enfermidade.

  Essa observação aparentemente banal indica que em muitos pacientes o sistema nervoso também é afetado pelo novo coronavírus.

* <u>**Cérebro**</u>

  A Mers e Sars, as infecções anteriores por coronavírus, já acusavam uma penetração dos vírus no cérebro, através dos nervos.

  Assim, pesquisadores do Japão e da China temem que, em alguns indivíduos, o patógeno penetre no tronco encefálico, danificando o centro respiratório lá localizado. Ainda não está claro se o Sars-cov-2 causa ou propicia acidentes vasculares cerebrais (AVC).

* <u>**Rins**</u>

  Pacientes de covid-19 com pneumonia que necessitem respiração artificial também podem sofrer lesões dos rins, sendo frequente a falência renal aguda.

  Além disso, o sangue coagula mais rápido em casos graves de covid-19, favorecendo a formação de coágulos que bloqueiam tanto os vasos sanguíneos quanto os rins. Têm sido observados pequenos infartos do tecido renal em diversos pacientes.

  Em cerca de 30% desses pacientes, a restrição das funções renais é tão aguda que eles necessitam de diálise. Ainda não está esclarecido se os rins saram após a recuperação da covid-19 ou se as lesões provocadas pelo Sars-cov-2 são duradouras.

* <u>**Pele**</u>

  O novo coronavírus igualmente causa danos visíveis ao maior órgão do corpo humano, a pele. Sobretudo crianças e jovens apresentaram pequenas lesões roxas nos pés, semelhantes às do sarampo, catapora ou frieiras.

  Nos artelhos (articulação), elas lembram úlceras de frio ou formam padrões reticulares, normalmente causados por coágulos nos capilares sanguíneos. Foram também observadas marcas, vermelhidão e urticária em outras partes do corpo. É possível que o tom azulado da pele se deva a uma coagulação patológica do sangue causada pelo novo coronavírus.

## 2.1- Dicionário de Variáveis

  Neste [Link](https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data-codebook.md), também no GitHub, é possível ver o Livro de códigos das **34 variáveis** deste *dataset* com suas resperctivas descrições e fontes extraídas.
  
  Detalhei as varíaveis (colunas) do *dataset* apenas com suas descrições traduzidas para nosso idioma conforme abaixo:
  
  `iso-code` = ISO 3166-1 alfa-3 - códigos de países com três letras;

  `continent`= Continente da localização geográfica;

  `location`= Localização geográfica;

  `date`= Data da observação;

  `total_cases`= Total de casos confirmados de COVID-19;

  `new_cases`= Novos casos confirmados de COVID-19;

  `total_deaths`= Total de mortes atribuídas ao COVID-19;

  `new_deaths`= Novas mortes atribuídas ao COVID-19;

  `total_cases_per_million` = Total de casos confirmados de COVID-19 por 1.000.000 de pessoas;

  `new_cases_per_million` = Novos casos confirmados de COVID-19 por 1.000.000 de pessoas;
  
  `total_deaths_per_million` = Total de mortes atribuídas ao COVID-19 por 1.000.000 de pessoas;

  `new_deaths_per_million` =	Novas mortes atribuídas ao COVID-19 por 1.000.000 de pessoas;

  `total_tests` = Total de testes para COVID-19;
  
  `new_tests`= Novos testes para COVID-19;

  `new_tests_smoothed` = Novos testes para COVID-19 (suavizado por 7 dias). Para países que não relatam dados de teste diariamente, assumimos que o teste mudou igualmente diariamente em qualquer período em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que são calculados como média em uma janela de 7 dias consecutivos;

  `total_tests_per_thousand` = Total de testes para COVID-19 por 1.000 pessoas;

  `new_tests_per_thousand` = Novos testes para COVID-19 por 1.000 pessoas;
  
  `new_tests_smoothed_per_thousand` =	Novos testes para COVID-19 (7 dias suavizados) por 1.000 pessoas;

  `tests_units` =	Unidades usadas pelo local para relatar seus dados de teste;

  `stringency_index` =	Índice de Estresse da Resposta do Governo: medida composta baseada em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibições de viagens, redimensionada para um valor de 0 a 100 (100 = resposta mais estrita);

  `population` =	População em 2020;

  `population_density` =	Número de pessoas divididas por área terrestre, medida em quilômetros quadrados, no ano mais recente disponível;

  `median_age` =	Idade média da população, projeção da ONU para 2020;

  `aged_65_older` =	Proporção da população com 65 anos ou mais, ano mais recente disponível;

  `ged_70_older` =	Percentagem da população com 70 anos ou mais em 2015;

  `gdp_per_capita` =	Produto interno bruto com paridade do poder de compra (dólares internacionais constantes de 2011), último ano disponível;
  
  `extreme_poverty` =	Percentagem da população que vive em extrema pobreza, último ano disponível desde 2010;
  
  `cvd_death_rate` =	Taxa de mortalidade por doenças cardiovasculares em 2017;
  
  `diabetes_prevalence` =	Prevalência de diabetes (% da população entre 20 e 79 anos) em 2017;
  
  `female_smokers`= Percentagem de mulheres que fumam, último ano disponível;
  
  `male_smokers`=	Proporção de homens que fumam, último ano disponível;

  `handwashing_facilities` =	Percentagem da população com instalações básicas de lavagem das mãos nas instalações, último ano disponível;
  
  `hospital_beds_per_thousand` =	Camas de hospital por 1.000 pessoas, último ano disponível desde 2010
  
  `life_expectancy` =	Expectativa de vida no nascimento em 2019.

  Para termos uma visão mais clara dessas informações, plotaremos um gráfico de barras simples.

```
# plotando gráfico de barras simples TOTAL DE MORTES
top_5_total_cases = df.loc[df.date == '2020-07-17', ['location', 'total_cases']].sort_values(by="total_cases", ascending=False)[1:6]
top_5_total_deaths = df.loc[df.date == '2020-07-17', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(6.95,3))
ax.bar('location', 'total_cases', data = top_5_total_cases)
ax.set_title ('Total de Casos')

fig, ax1 = plt.subplots(figsize=(6,3))
ax1.bar('location', 'total_deaths', data = top_5_total_deaths)
ax1.set_title ('Total de Mortes')

plt.tight_layout();
```

  Encerraremos a primeira fase da análise plotando dois gráficos, sendo o primeiro para mostrar como está a evolução do `total_cases` (Total de Casos) no Mundo e o segundo um gráfico de disperção considerando o `gpd_per_capita` (PIB dos países) em relação ao `total_deaths` (Total de Mortos).

```
# Gráfico 01 - Gráfico de linhas para ver a evolução do total de mortes de COVID-19 no mundo
df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot();
```

```
# Gráfico 02 - Gráfico de Disperção Número de Casos no Mundo
fig, ax = plt.subplots(figsize=(10,8))

# plotar gpd_per_capita vs. total_deaths
df.plot.scatter(x='gdp_per_capita', y='total_cases', ax=ax, s=5)
ax.set_title('Relação PIB x Total de Casos')
ax.set_xlabel('PIB')
ax.set_ylabel('Total de Casos')

plt.show()
```

  Conforme os gráficos acima, o **Gráfico 01** não mostra um cenário muito animador até a atual data (17-07-2020), pois, a linha mostra uma ascendência constante.

  No **Gráfico 02**, percebemos que o número maior de casos está concentrado em países que têm seu PIB entre 10.000 e 20.000.

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
#df_brasil = df.loc[df.location == 'Brazil'].copy()
x = list(range(200))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15, 9))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica)')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()  
```

  Os gráficos acima, mostram que tanto a quantidade de casos quanto a quantidades de mortes sobem de maneira vertiginosa.
  Nas escalas logarítmas (gráficos à esquerda), mostra uma curva bastante acentuada, porém, com ritmo ascendente, ou seja, ainda estamos numa posição que a prevenção e o isolamento social ainda são as únicas medidas eficazes até o surgimento de uma vacina.
